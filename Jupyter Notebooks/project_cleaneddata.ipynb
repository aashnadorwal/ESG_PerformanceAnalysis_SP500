```python
import pandas as pd

# ESG data
esg_df = pd.read_csv("sp500_esg_data.csv")
# S&P 500 companies data
companies_df = pd.read_csv("sp500_companies.csv")

print("ESG Data:")
display(esg_df.head())

print("Companies Data:")
display(companies_df.head())
```

```python
esg_columns = {
    "Full Name": "Company Name",
    "Symbol": "Ticker Symbol",
    "GICS Sector": "Sector",
    "totalEsg": "Total ESG Score",
    "environmentScore": "Environmental Score",
    "socialScore": "Social Score",
    "governanceScore": "Governance Score",
    "highestControversy": "Highest Controversy",
    "overallRisk": "Overall Risk",
}
esg_df = esg_df[list(esg_columns.keys())].rename(columns=esg_columns)
esg_df
```

```python
company_columns = {
    "Symbol": "Ticker Symbol",
    "Fulltimeemployees": "Full-time Employees",
    "Marketcap": "Market Cap",
    "Revenuegrowth": "Revenue Growth"
}
companies_df = companies_df[list(company_columns.keys())].rename(columns=company_columns)
companies_df.head()
```

```python
# Integrating ESG data with Companies data by Ticker Symbol
merged_df = esg_df.merge(companies_df, on="Ticker Symbol", how="inner")
merged_df
```

```python
missing_values = merged_df.isna().sum()
missing_values
```

```python
merged_df.dropna(inplace=True)
merged_df
```

```python
merged_df.to_csv("sp500_bothmerged.csv", index=False)
merged_df
```

```python
# Merging web scraped SlickCharts df to Merged df
slickcharts_df = pd.read_csv("sp500_slickcharts.csv")
slickcharts_df
```

```python
slickcharts_df.rename(columns={"Symbol": "Ticker Symbol"}, inplace=True)

final_df = merged_df.merge(slickcharts_df[["Ticker Symbol", "YTD Return"]], on="Ticker Symbol", how="inner")
final_df
```

```python
final_df.to_csv("final_merged_sp500.csv", index=False)
```
